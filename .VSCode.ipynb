```python
#after compliting EDA and feature engineering part and now start the feature selection part
dff=pd.read_csv("cleand_ridge_lasso_data.csv")
dff.columns
```

```python
dff.head()
```

```python
dff.drop(["day","month","year"],axis=1,inplace=True)
```

```python
#indipendent and dependent feature
x=dff.drop("FWI",axis=1)
y=dff["FWI"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
#x_train correlation
x_train.corr()
```

```python
sns.heatmap(x_train.corr(),annot=True)
```

```python
#feature standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
#boxplot to understand effect of standard scaler
plt.subplot(1,2,1)
plt.boxplot(x_train)
plt.title("before stand scaler")
plt.subplot(1,2,2)
plt.boxplot(x_train_scaled)
plt.title("after stand scaler")
```

```python
#model traing
from sklearn.linear_model import LinearRegression
linear=LinearRegression()
linear.fit(x_train_scaled,y_train)
y_pred=linear.predict(x_test_scaled)
from sklearn.metrics import mean_absolute_error,r2_score
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print(f"mae : {mae}")
print(f"r2_score : {score}")
```

```python
plt.scatter(y_test,y_pred)
```

```python
#applying Lasso regression
from sklearn.linear_model import Lasso
la=Lasso()
la.fit(x_train_scaled,y_train)
y_pred=la.predict(x_test_scaled)
from sklearn.metrics import mean_absolute_error,r2_score
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print(f"mae : {mae}")
print(f"r2_score : {score}")
```

```python
plt.scatter(y_test,y_pred)
```

```python
#ridge regression
from sklearn.linear_model import Ridge
ri=Ridge()
ri.fit(x_train_scaled,y_train)
y_pred=ri.predict(x_test_scaled)
from sklearn.metrics import mean_absolute_error,r2_score
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print(f"mae : {mae}")
print(f"r2_score : {score}")
```
